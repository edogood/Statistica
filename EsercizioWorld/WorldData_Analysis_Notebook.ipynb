```

# Jupyter Notebook: Analisi esplorativa del dataset WorldData con grafici

# Import delle librerie necessarie
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Caricamento del dataset
df = pd.read_csv('WorldData_CLEAN 1.csv')

# Pulizia dei dati (rimozione dei valori mancanti)
df_clean = df.dropna()

# Grafico della distribuzione della popolazione totale
plt.figure(figsize=(10,6))
sns.histplot(df_clean['Population_Total'], bins=30, kde=True)
plt.xlabel('Popolazione Totale')
plt.ylabel('Numero di Paesi')
plt.title('Distribuzione della Popolazione Totale')
plt.grid(False)
plt.show()

# Grafico della distribuzione della densità di popolazione
plt.figure(figsize=(10,6))
sns.histplot(df_clean['Density
(P/Km2)'], bins=30, kde=True, color='orange')
plt.xlabel('Densità della Popolazione (P/Km²)')
plt.ylabel('Numero di Paesi')
plt.title('Distribuzione della Densità della Popolazione')
plt.grid(False)
plt.show()

# Grafico della distribuzione delle emissioni di CO2
plt.figure(figsize=(10,6))
sns.histplot(df_clean['Co2_Emissions'], bins=30, kde=True, color='green')
plt.xlabel('Emissioni di CO2')
plt.ylabel('Numero di Paesi')
plt.title('Distribuzione delle Emissioni di CO2')
plt.grid(False)
plt.show()

# Matrice di correlazione visualizzata come heatmap
plt.figure(figsize=(12,8))
sns.heatmap(df_clean.corr(), annot=True, cmap='coolwarm')
plt.title('Matrice di Correlazione')
plt.show()

# Scatter plot delle emissioni di CO2 vs popolazione urbana
plt.figure(figsize=(10,6))
sns.scatterplot(x='Population_Urban', y='Co2_Emissions', data=df_clean)
plt.xlabel('Popolazione Urbana')
plt.ylabel('Emissioni di CO2')
plt.title('Emissioni di CO2 vs Popolazione Urbana')
plt.grid(True)
plt.show()

# Scatter plot delle emissioni di CO2 vs area agricola
plt.figure(figsize=(10,6))
sns.scatterplot(x='Agricultural_Land(Km2)', y='Co2_Emissions', data=df_clean, color='purple')
plt.xlabel('Area Agricola (Km²)')
plt.ylabel('Emissioni di CO2')
plt.title('Emissioni di CO2 vs Area Agricola')
plt.grid(True)
plt.show()

# Scatter plot della densità di popolazione vs area forestale
plt.figure(figsize=(10,6))
sns.scatterplot(x='Forested_Area (Km2)', y='Density
(P/Km2)', data=df_clean, color='brown')
plt.xlabel('Area Forestale (Km²)')
plt.ylabel('Densità di Popolazione (P/Km²)')
plt.title('Densità di Popolazione vs Area Forestale')
plt.grid(True)
plt.show()
```
